# Convolutional Neural Network
## Classifying whether the image is of a cat or a dog.

```python
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
training_set = image_dataset_from_directory(
  r"Training-set-path",
  image_size = (64,64),
  batch_size = 32
)
```

```python
testing_set = image_dataset_from_directory(
    r"Test-set-path",
    image_size=(64,64),
    batch_size = 32
)
```

```python
normalization = tf.keras.layers.Rescaling(1./255)

data_augmentation = tf.keras.models.Sequential([
  tf.keras.layers.RandomFlip("horizontal"),
  tf.keras.layers.RandomRotation(0.1),
  tf.keras.layers.RandomZoom(0.1)
])
```

```python
cnn = tf.keras.Sequential([normalization,data_augmentation])
```

### Convolution Layer

```python
cnn.add(tf.keras.layers.Conv2D(filters = 32,kernel_size=3,activation='relu',input_shape = [64,64,3]))
```

### Pooling


```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

### Adding a new Convolutional Layer

```python
cnn.add(tf.keras.layers.Conv2D(filters = 32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

### Adding a flattening layer

```python
cnn.add(tf.keras.layers.Flatten())
```

### Adding Fully Connected Layer

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

### Adding the output layer

```python
cnn.add(tf.keras.layers.Dense(units=1,activation = "sigmoid"))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x = training_set,validation_data=testing_set,epochs = 25)
```

```python
import numpy as np
from keras.preprocessing import image

test_image = image.load_img(r"C:\Users\vijay\Downloads\Deep Learning A-Z\Part 2 - Convolutional Neural Networks (CNN)\dataset\single_prediction\cat_or_dog_2.jpg",target_size = (64,64))
img_array = image.img_to_array(test_image)
predict_image = np.expand_dims(img_array,axis=0)
predictions = cnn.predict(predict_image)
if predictions[0][0] == 1:
  print("Dog")
else:
  print("Cat")
```
