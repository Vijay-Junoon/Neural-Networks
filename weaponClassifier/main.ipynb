```python
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
train_set = image_dataset_from_directory(
  r"C:\Users\vijay\Downloads\archive (29)\weapon_detection\train\images",
  image_size = (64,64),
  batch_size = 32
)

classes_info = train_set.class_names
n_classes = len(classes_info)
```

```python
validation_set = image_dataset_from_directory(
  r"C:\Users\vijay\Downloads\archive (29)\weapon_detection\val\images",
  image_size = (64,64),
  batch_size = 32
)
```

```python
normalization = tf.keras.layers.Rescaling(1./255)
```

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.1),
  tf.keras.layers.RandomZoom(0.1)
])
```

```python
cnn = tf.keras.Sequential([normalization,data_augmentation])
```

```python
cnn.add(tf.keras.layers.Conv2D(kernel_size=3,filters=32,activation='relu',input_shape=(64,64,3)))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(kernel_size = 3,filters = 64,activation = 'relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(kernel_size=3,filters = 128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=n_classes,activation='softmax'))
```

```python
cnn.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x=train_set,validation_data=validation_set,epochs=35)
```

```python
classes_map = {i: classes_info[i] for i in range(len(classes_info))}
import numpy as np

img = tf.keras.utils.load_img('test-1.jpeg',target_size = (64,64))
img_array = tf.keras.utils.img_to_array(img)
img_array = np.expand_dims(img_array,0)
predictions = cnn.predict(img_array)


print(classes_map[np.argmax(predictions[0])])
```
