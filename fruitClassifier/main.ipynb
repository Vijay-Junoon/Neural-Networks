```python
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
train_dataset = image_dataset_from_directory(
  r"Training-set-path",
  image_size = (64,64),
  batch_size = 32
)
```

```python
validation_dataset = image_dataset_from_directory(
  r"Validation-set-path",
  image_size = (64,64),
  batch_size = 32
)
```

### Normalizing

```python
normalization = tf.keras.layers.Rescaling(1./255)
```

### Data Augmentation

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.1),
  tf.keras.layers.RandomZoom(0.1)
])
```

### CNN - Initialization

```python
cnn = tf.keras.Sequential([normalization,data_augmentation])
```

### Adding Convolutional Layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=  32,kernel_size=3,activation = 'relu',input_shape = (64,64,3)))
```

### Add Pooling Layer

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

### Add 2nd convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters = 64,kernel_size=3,activation = 'relu'))
```

### Add 2nd Pooling layer

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2,strides = 2))
```

### Adding a Flattening Layer

```python
cnn.add(tf.keras.layers.Flatten())
```

### Adding a fully connected layer

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

### Adding the output layer

```python
n_units = len(train_dataset.class_names)
cnn.add(tf.keras.layers.Dense(units=n_units,activation = 'softmax'))
```

### Compiling the CNN

```python
cnn.compile(optimizer = 'adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

### Training the CNN

```python
cnn.fit(x=train_dataset,validation_data=validation_dataset,epochs=25)
```

### Testing the model

```python
import numpy as np

img = tf.keras.utils.load_img(r'test_img.jpg',target_size = (64,64))
img_array = tf.keras.utils.img_to_array(img)

img_array = np.expand_dims(img_array,0)

predictions = cnn.predict(img_array)

classes =  {i : train_dataset.class_names[i] for i in range(len(train_dataset.class_names))}

print(classes[np.argmax(predictions[0])])

```
